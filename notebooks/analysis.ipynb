```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../output/data/gold/filtred_with_coordinates.xlsx')
df = df.loc[(df['preco']>49999)&(df['area util por m²']>9)]
```

```python
to_chart = df.loc[df['cep_dest']=='Centro'][['bairro','latitude_origin','longitude_origin']].reset_index(drop=True)

center_lat = to_chart['latitude_origin'].mean()
center_lon = to_chart['longitude_origin'].mean()

fig = px.scatter_mapbox(to_chart,
    lat='latitude_origin',
    lon='longitude_origin',
    hover_name='bairro',
    hover_data={'bairro': True},
    color_discrete_sequence=['red'],
    zoom=10,
    width=500,
    height=650,
    mapbox_style="open-street-map",
    center={"lat": center_lat+0.035, "lon": center_lon}
)

fig.update_layout(
    margin={"r":0, "t":0, "l":0, "b":0},
    title="Distribuição de Pontos de Origem"
)

fig.show()
```

# Todas as zonas do município de Ilhéus

```python
all_zonas = df.loc[df['cep_dest'] == 'Centro']['Zona'].value_counts()
df_all_zonas = df.loc[df['Zona'].isin(all_zonas.index)][['Zona']].drop_duplicates(subset='Zona').reset_index(drop=True)
df_all_zonas['count'] = df_all_zonas['Zona'].map(all_zonas)

total_count = df_all_zonas['count'].sum()

df_all_zonas['pct'] = df_all_zonas['count'] / total_count * 100
df_all_zonas['pct_str'] = df_all_zonas['pct'].apply('{:.1f}%'.format)

df_all_zonas['pct'] = df_all_zonas['count'] / total_count * 100
df_all_zonas['pct_str'] = df_all_zonas['pct'].apply('{:.1f}%'.format)

df_all_zonas = df_all_zonas.sort_values(by='count', ascending=True)

fig = px.bar(df_all_zonas, x='count', y='Zona', orientation='h', text='pct_str',
             text_auto=False,
             labels={'count': 'Apartamentos à Venda'})

fig.update_traces(textangle=0, textposition="outside", textfont_size=12)

fig.update_layout(
    margin=dict(l=0, r=20, t=0, b=0),
    height=500,
    width=800,
    xaxis_title="", 
    yaxis_title="",
    xaxis_range=[0, 1000])

fig.show()
```

```python
all_zonas = df.loc[df['cep_dest'] == 'Centro']

metrics_all_zones = pd.DataFrame(columns=['Zona', 'Quantidade de imoveis', 'Preço médio', 'área util por m2', 'Preço por m2'])

for zona in all_zonas['Zona'].unique():
    quantidade_imoveis = np.round(len(all_zonas.loc[all_zonas['Zona'] == zona]), 2)
    preco_medio = np.round(all_zonas.loc[all_zonas['Zona'] == zona]['preco'].mean(), 2)
    area_util_por_m2 = np.round(all_zonas.loc[all_zonas['Zona'] == zona]['area util por m²'].mean(), 2)
    preco_por_m2 = np.round(preco_medio / area_util_por_m2, 2)

    line = pd.DataFrame({'Zona': [zona],
                               'Quantidade de imoveis': [quantidade_imoveis],
                               'Preço médio': [preco_medio],
                               'área util por m2': [area_util_por_m2],
                               'Preço por m2': [preco_por_m2]})

    metrics_all_zones = pd.concat([metrics_all_zones, line], ignore_index=True)

metrics_all_zones
```

## Descrição dos bairros da zona sul

```python
somente_zona_sul = df.loc[df['Zona'] == 'Sul']
somente_zona_sul = somente_zona_sul.loc[somente_zona_sul['cep_dest'] == 'Centro']

df_resultado = pd.DataFrame(columns=['Bairro', 'Quantidade de imoveis', 'Preço médio', 'área util por m2', 'Preço por m2'])

for bairro in somente_zona_sul['bairro'].unique():
    quantidade_imoveis = np.round(len(somente_zona_sul.loc[somente_zona_sul['bairro'] == bairro]), 2)
    preco_medio = np.round(somente_zona_sul.loc[somente_zona_sul['bairro'] == bairro]['preco'].mean(), 2)
    area_util_por_m2 = np.round(somente_zona_sul.loc[somente_zona_sul['bairro'] == bairro]['area util por m²'].mean(), 2)
    preco_por_m2 = np.round(preco_medio / area_util_por_m2, 2)

    nova_linha = pd.DataFrame({'Bairro': [bairro],
                               'Quantidade de imoveis': [quantidade_imoveis],
                               'Preço médio': [preco_medio],
                               'área util por m2': [area_util_por_m2],
                               'Preço por m2': [preco_por_m2]})

    df_resultado = pd.concat([df_resultado, nova_linha], ignore_index=True)

df_resultado
```

```python
cols_caract_apt = somente_zona_sul[['bairro', 'quartos', 'banheiros', 'vagas garagem','custo condominio', 'iptu']].replace(-1, 0)

df_resultado = pd.DataFrame(columns=['Bairro', 'Media quartos', 'Media banheiros', 'Media garagem', 'Media custo condominio', 'Media iptu'])

for bairro in cols_caract_apt['bairro'].unique():
    df_for_this_bairro = cols_caract_apt[cols_caract_apt['bairro'] == bairro]
    media_quartos = df_for_this_bairro['quartos'].mean()
    media_banheiros = df_for_this_bairro['banheiros'].mean()
    media_garagem = df_for_this_bairro['vagas garagem'].mean()
    media_condominio = df_for_this_bairro['custo condominio'].mean()
    media_iptu = df_for_this_bairro['iptu'].mean()

    nova_linha = pd.DataFrame({'Bairro': [bairro],
                               'Media quartos': [media_quartos],
                               'Media banheiros': [media_banheiros],
                               'Media garagem': [media_garagem],
                               'Media custo condominio': [media_condominio],
                               'Media iptu': [media_iptu]})

    df_resultado = pd.concat([df_resultado, nova_linha], ignore_index=True)
```

```python
cols_caract_cond = somente_zona_sul[['bairro', 'armários na cozinha', 'academia', 'piscina', 'varanda', 'portaria',
                                     'quarto de serviço', 'área de serviço', 'condomínio fechado', 'permitido animais', 
                                     'churrasqueira', 'elevador', 'mobiliado','armários no quarto', 'ar condicionado',
                                     'segurança 24h','salão de festas']]

final_result = {}
for bairro in cols_caract_cond['bairro'].unique():
    intermedian_result = dict()
    cols_to_sumarize = [cols for cols in cols_caract_cond if cols != 'bairro']
    for col in cols_to_sumarize:
        media = (cols_caract_cond.loc[cols_caract_cond['bairro'] == bairro][col].values.sum() /
                 len(cols_caract_cond.loc[cols_caract_cond['bairro'] == bairro])) * 100
        intermedian_result[col] = np.round(media,2)
    final_result[bairro] = intermedian_result

df_final_result_from_cols_caract_cond = pd.DataFrame(final_result)
df_final_result_from_cols_caract_cond
```

```python
fig = px.imshow(df_final_result_from_cols_caract_cond,
                labels=dict(x="Bairros", y="Características", color="Porcentagem"),
                x=df_final_result_from_cols_caract_cond.columns,
                y=df_final_result_from_cols_caract_cond.index,
                color_continuous_scale='Viridis',
                text_auto=True)

fig.update_xaxes(title_text='')
fig.update_yaxes(title_text='')

fig.update_layout(xaxis_nticks=36,
                  width=1000,
                  height=800)

fig.show()
```
